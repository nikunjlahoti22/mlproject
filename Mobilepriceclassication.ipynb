```python
X=df.drop('price_range',axis=1)

y=df['price_range']

```

**Splitting the data**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Training with LinearRegression

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

lm.fit(X_train,y_train)
lm.score(X_test,y_test)
```

Creating and training KNN model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

Creating and training LoogisticRegression

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()

logmodel.fit(X_train,y_train)
logmodel.score(X_test,y_test)
```

Training with DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()

dtree.fit(X_train,y_train)
dtree.score(X_test,y_test)
```

Creating and training RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, y_train)
rfc.score(X_test,y_test)
```

**Conclusion-**LinearRegreesion and KNN performed best

```python
# Result using linear Regression
y_pred=lm.predict(X_test)
```

```python
# Result using KNN
from sklearn.metrics import classification_report,confusion_matrix
pred = knn.predict(X_test)
print(classification_report(y_test,pred))
```

```python
matrix=confusion_matrix(y_test,pred)
print(matrix)
```
